```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.preprocessing import LabelEncoder
from collections import Counter
```

```python
df = pd.read_csv('zomato.csv')
```

```python
null_values =df.isnull().mean().mul(100)
# null_values.sort()
print(null_values.sort_values())
```

```python
df.drop(['url','address','dish_liked','phone','menu_item','reviews_list','listed_in(city)'],axis =1 ,inplace = True)
```

```python
df.info()
```

```python
le_onlineoreder = LabelEncoder()
le_booktable = LabelEncoder()
df["online_orderN"] = le_onlineoreder.fit_transform(df['online_order'])
df["book_tableN"] = le_booktable.fit_transform(df['book_table'])
df.drop(columns=['online_order'],inplace=True)
df.drop(columns=['book_table'],inplace=True)
df.head(100)
```

```python
null_values =df.isnull().mean().mul(100)
# null_values.sort()
print(null_values.sort_values())
```

```python
df.dtypes


```

```python
df.head(100)
```

```python
df['rate'] = df['rate'].str.replace('/5', '')  
df['rate'] = pd.to_numeric(df['rate'],errors='coerce')

df.head(10)
```

```python
g = df.groupby('approx_cost(for two people)')
g.groups.keys()
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',','')
# df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(int)
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'],errors='coerce')

```

```python
g = df.groupby('listed_in(type)')
g.groups.keys()
```

```python
df['Buffet'] = np.where(df['listed_in(type)']=='Buffet', 1, 0)
df['Cafes'] = np.where(df['listed_in(type)']=='Cafes', 1, 0)
df['Delivery'] = np.where(df['listed_in(type)']=='Delivery', 1, 0)
df['Desserts'] = np.where(df['listed_in(type)']=='Desserts', 1, 0)
df['Dine-out'] = np.where(df['listed_in(type)']=='Dine-out', 1, 0)
df['Drinks & nightlife'] = np.where(df['listed_in(type)']=='Drinks & nightlife', 1, 0)
df['Pubs and bars'] = np.where(df['listed_in(type)']=='Pubs and bars', 1, 0)
```

```python
df.drop(columns=['listed_in(type)'],inplace=True)
```

```python
df.dtypes
```

```python
df['location'] = df['location'].fillna('Unknown')
df['rest_type'] = df['rest_type'].fillna('Unknown')
df['cuisines']  = df['cuisines'].fillna('Unknown')
df['approx_cost(for two people)']  = df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].median())
```

```python
null_values =df.isnull().mean().mul(100)
# null_values.sort()
print(null_values.sort_values())
```

```python
# Frequency encode location
location_freq = df['location'].value_counts(normalize=True)
df['location'] = df['location'].map(location_freq)

# Frequency encode rest_type
rest_type_freq = df['rest_type'].value_counts(normalize=True)
df['rest_type'] = df['rest_type'].map(rest_type_freq)
```

```python
df.head()
```

```python
df['cuisines_list'] = df['cuisines'].str.split(', ')
all_cuisines = df['cuisines_list'].explode()
cuisine_counts = Counter(all_cuisines)
top_25 = [cuisine for cuisine, count in cuisine_counts.most_common(25)]
print(top_25)
```

```python
def keep_top_cuisines(cuisine_list):
    return [c if c in top_25 else 'Other' for c in cuisine_list]

df['cuisines_list'] = df['cuisines_list'].apply(keep_top_cuisines)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
cuisine_encoded = mlb.fit_transform(df['cuisines_list'])

# convert to dataframe with proper column names
cuisine_df = pd.DataFrame(
    cuisine_encoded,
    columns=['cuisine_' + c for c in mlb.classes_],
    index=df.index
)

print(cuisine_df.shape)   # should be (41263, 26) — 25 cuisines + Other
print(cuisine_df.columns.tolist())
```

```python
df = pd.concat([df, cuisine_df], axis=1)
df.drop(columns=['cuisines', 'cuisines_list'], inplace=True)
```

```python
df.head()
```

```python
df.dropna(subset=["rate"])
df.drop(columns= ['name'],inplace=True)
```

```python
df.dtypes
```

```python
df = df.replace(
    ['nan', 'NaN', 'NAN', 'none', 'None', 'NONE', 
     'null', 'Null', 'NULL', 'na', 'NA', 'N/A', 'Nan',
     '', ' ', '-', '--', '?'],
    np.nan
)

# Remove spaces from string columns
df = df.apply(lambda col: col.str.strip() if col.dtype == 'object' else col)

# Again convert empty strings after stripping
df = df.replace('', np.nan)

# Drop all rows that have even one missing value
df = df.dropna()

print(df.isnull().sum())
print(df.shape)
```

```python
df.to_csv('final.csv', index=False)
```
